["Deepfakes, Elections, and Shrinking the Liar's Dividend]https://www.brennancenter.org/our-work/research-reports/deepfakes-elections-and-shrinking-liars-dividend

This article mainly explains about how the existence of, and education about,  AI affects accountability. They call this effect the Liar’s Dividend and talk about both the way it is used and the strategies in its use. They explain the Liar’s dividend through how it is used by politicians and with 2 axes with how it worked, distance to the person making the claim and the directness of the statement. They concluded that the main way to reduce these claims is to increase the accountability for people who are lying. 

[Reality Is Melting as Lawyers Claim Real Videos Are Deepfakes]https://futurism.com/reality-melting-lawyers-deepfakes

This article explains how A makes it possible to create uncertainty in court records. It explains that due to the increased alertness of various ways to fake events a so-called “deepfake defense” has spread. By claiming that real evidence is fake it is possible to sway juror opinion by creating an altered reality. The article ends with a warning of the future of the possible spread of the deepfake defense due to its nature of being a gray area in the law.
